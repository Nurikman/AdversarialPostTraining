```python
import pandas as pd

csv_path = "/Users/nursultanatymtay/Desktop/Algoverse 2026/AdversarialPostTraining/openai_sft_900/gsm8k_eval/gsm8k_accuracy_summary_fine_tune_run_20260519T134603Z_train_4001_4100.csv"
df = pd.read_csv(csv_path)

df
```

```python
import matplotlib.pyplot as plt

ax = df.plot(
    x="model_label",
    y="correct",
    kind="bar",
    legend=False,
    figsize=(8, 5),
    color="skyblue"
)

ax.set_title("Correct Answers by Model")
ax.set_xlabel("Model Label")
ax.set_ylabel("Correct")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
